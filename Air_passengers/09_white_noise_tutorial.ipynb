### In discrete time, white noise is a discrete signal whose samples are regarded as a sequence of serially uncorrelated random variables with zero mean and finite variance; a single realization of white noise is a random shock.

```python

from random import gauss
from random import seed
from pandas import Series
from pandas.plotting import autocorrelation_plot
# seed random number generator
seed(1)
# create white noise series
series = [gauss(0.0, 1.0) for i in range(1000)]
series = Series(series)
```

```python
series
```

```python
series.hist()
```

```python
series.describe()
```

## mean =0 and std = 1

```python
# line plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
series.plot()
plt.show()
```

```python
# autocorrelation
plt.figure(figsize=(10,8))
autocorrelation_plot(series)
plt.show()
```

## White noise time series is defined by a zero mean, constant variance, and zero correlation.
